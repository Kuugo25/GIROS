```python
import copy
from simulation.artifact import simulate_artifact
from core.talents import load_talent_multipliers
from core.utils import make_clean_character

artifacts = {
    "Flower": {
        "main_stat": "Flat HP",
        "main_stat_value": 4780,  # Standard flower main stat
        "substats": {
            "CRIT Rate%": 10.5,
            "CRIT DMG%": 30.0,
            "HP%": 5.8,
            "ATK%": 5.8
        },
        "upgrades": 5
    },
    "Feather": {
        "main_stat": "Flat ATK",
        "main_stat_value": 311,  # Standard feather main stat
        "substats": {
            "CRIT Rate%": 10.5,
            "CRIT DMG%": 30.0,
            "HP%": 5.8,
            "ATK%": 5.8
        },
        "upgrades": 5
    },
    "Sands": {
        "main_stat": "HP%",
        "main_stat_value": 46.6,  # Typical HP% sands main stat
        "substats": {
            "CRIT Rate%": 10.5,
            "CRIT DMG%": 30.0,
            "Flat HP": 298,
            "ATK%": 5.8
        },
        "upgrades": 5
    },
    "Goblet": {
        "main_stat": "Pyro DMG Bonus",
        "main_stat_value": 46.6,
        "substats": {
            "CRIT Rate%": 10.5,
            "CRIT DMG%": 30.0,
            "HP%": 5.8,
            "ATK%": 5.8
        },
        "upgrades": 5
    },
    "Circlet": {
        "main_stat": "CRIT DMG%",
        "main_stat_value": 62.2,
        "substats": {
            "CRIT Rate%": 23.0,
            "HP%": 5.8,
            "ATK%": 5.8,
            "Flat HP": 298
        },
        "upgrades": 5
    }
}
def test_artifact_replacement(
    character,
    slot="Goblet",
    multipliers=None,
    combo="E12N1C",
    skill="AA"
):
    # 1) Load multipliers if needed
    if multipliers is None:
        multipliers = load_talent_multipliers("../data/talent_multipliers.csv")

    # 2) Baseline copy & DPS
    char_before = copy.deepcopy(character)
    char_before.sync_stats(char_before.compute_total_stats())
    dmg_before = char_before.expected_damage_output(
        combo=combo,
        skill=skill,
        multipliers=multipliers,
        crit_rate=char_before.crit_rate,
        crit_dmg=char_before.crit_dmg,
        dmg_bonus=char_before.dmg_bonus,
        enemy_level=100,
        enemy_resistance=0.1
    )
    stats_before = char_before.compute_total_stats()

    # 3) Roll new artifact and test on a fresh copy
    new_art = simulate_artifact(slot)
    char_after = copy.deepcopy(character)
    char_after.artifacts[slot] = new_art
    char_after.sync_stats(char_after.compute_total_stats())

    dmg_after = char_after.expected_damage_output(
        combo=combo,
        skill=skill,
        multipliers=multipliers,
        crit_rate=char_after.crit_rate,
        crit_dmg=char_after.crit_dmg,
        dmg_bonus=char_after.dmg_bonus,
        enemy_level=100,
        enemy_resistance=0.1
    )
    stats_after = char_after.compute_total_stats()

    # 4) If it improves DPS, apply to original
    accepted = (dmg_after > dmg_before)
    if accepted:
        character.artifacts[slot] = new_art
        character.sync_stats(character.compute_total_stats())

    # 5) Print results
    delta   = dmg_after - dmg_before
    pct     = delta / dmg_before * 100 if dmg_before else 0
    sign    = "+" if delta>=0 else "-"

    print(f"\n=== Artifact Replacement Test ({slot}) ===")
    print(f"Old Damage: {dmg_before:.2f}")
    print(f"New Damage: {dmg_after:.2f}")
    print(f"Δ Damage: {sign}{abs(delta):.2f} ({sign}{abs(pct):.2f}%)\n")

    print("🧾 Before Stats:")
    print(f"  HP:         {stats_before['total_hp']:.2f}")
    print(f"  ATK:        {stats_before['total_atk']:.2f}")
    print(f"  DEF:        {stats_before['total_def']:.2f}")
    print(f"  CRIT Rate:  {stats_before['crit_rate']:.2%}")
    print(f"  CRIT DMG:   {stats_before['crit_dmg']:.2%}")
    print(f"  {character.vision} Bonus: {stats_before['elemental_dmg_bonus'][character.vision]:.2%}\n")

    print("🧾 After Stats:")
    print(f"  HP:         {stats_after['total_hp']:.2f}")
    print(f"  ATK:        {stats_after['total_atk']:.2f}")
    print(f"  DEF:        {stats_after['total_def']:.2f}")
    print(f"  CRIT Rate:  {stats_after['crit_rate']:.2%}")
    print(f"  CRIT DMG:   {stats_after['crit_dmg']:.2%}")
    print(f"  {character.vision} Bonus: {stats_after['elemental_dmg_bonus'][character.vision]:.2%}\n")

    print("New artifact rolled:", new_art)


# Example usage:
hu_tao = make_clean_character("Hu Tao", weapon_name="Homa", artifacts=artifacts)
test_artifact_replacement(hu_tao, slot="Goblet")
```

```python
from core.utils import make_clean_character
from core.talents import load_talent_multipliers
import copy

def test_goblet_vs_goblet(
    character_factory,
    multipliers,
    combo="EQ",
    slot="Goblet",
    main_stat_value=46.6,
    substats=None
):
    """
    Compare DPS for a Pyro DMG Bonus goblet vs a DEF% goblet,
    holding all substats constant.
    """
    if substats is None:
        substats = {
            "CRIT Rate%": 10.5,
            "CRIT DMG%": 30.0,
            "HP%": 5.8,
            "ATK%": 5.8
        }

    # Build the two artifact dicts
    pyro_goblet = {
        "main_stat": "HP%",
        "main_stat_value": main_stat_value,
        "substats": substats,
        "upgrades": 5
    }
    def_goblet = {
        "main_stat": "ATK%",
        "main_stat_value": main_stat_value,
        "substats": substats,
        "upgrades": 5
    }

    # Create fresh Hu Tao for each test
    base = character_factory()
    # Pyro‐bonus test
    hu_pyro = copy.deepcopy(base)
    hu_pyro.artifacts[slot] = pyro_goblet
    hu_pyro.sync_stats(hu_pyro.compute_total_stats())
    d_pyro = hu_pyro.expected_damage_output(
        combo=combo,
        skill=skill,
        multipliers=multipliers,
        enemy_level=100,
        enemy_resistance=0.1
    )

    # DEF% test
    hu_def = copy.deepcopy(base)
    hu_def.artifacts[slot] = def_goblet
    hu_def.sync_stats(hu_def.compute_total_stats())
    d_def = hu_def.expected_damage_output(
        combo=combo,
        skill=skill,
        multipliers=multipliers,
        enemy_level=100,
        enemy_resistance=0.1
    )

    delta = d_pyro - d_def
    pct = delta / d_def * 100 if d_def else 0

    print(f"→ HP%-Goblet DPS : {d_pyro:.2f}")
    print(f"→ ATK%-Goblet DPS: {d_def:.2f}")
    print(f"Δ DPS (pyro−def): {delta:.2f} ({pct:+.2f}%)")

    return {
        "pyro_dps": d_pyro,
        "def_dps": d_def,
        "delta": delta,
        "pct_diff": pct
    }

# 1. Load your multipliers & character factory
multipliers = load_talent_multipliers("../data/talent_multipliers.csv")
hu_tao_factory = lambda: make_clean_character("Hu Tao", weapon_name="Homa", artifacts=artifacts)

# 2. Call the comparison
res = test_goblet_vs_goblet(
    character_factory=hu_tao_factory,
    multipliers=multipliers,
    combo="EQ"
)
```

```python
import pandas as pd
from core.utils        import build_character_factory
from core.talents      import load_talent_multipliers
from core.summons      import HealEffect
from simulation.simulator import (
    simulate_with_metric,
    simulate_artifact_farm_heal,
    simulate_talent_farm_heal
)

# 0) Your artifact dict (biasing HP%/ER% is fine)
artifacts = {
    "Flower": {
        "main_stat": "Flat HP",
        "main_stat_value": 4780,
        "substats": {"HP%":5.8, "Energy Recharge%": 10.0},
        "upgrades": 5
    },
    "Feather": {
        "main_stat": "Flat HP",
        "main_stat_value": 311,
        "substats": {"HP%":5.8, "Energy Recharge%": 10.0},
        "upgrades": 5
    },
    "Sands": {
        "main_stat": "HP%",
        "main_stat_value": 46.6,
        "substats": {"Energy Recharge%":10.5, "HP%":5.8},
        "upgrades": 5
    },
    "Goblet": {
        "main_stat": "HP%",
        "main_stat_value": 46.6,
        "substats": {"Energy Recharge%":10.5, "HP%":5.8},
        "upgrades": 5
    },
    "Circlet": {
        "main_stat": "Healing Bonus%",
        "main_stat_value": 35.9,
        "substats": {"Energy Recharge%":10.5, "HP%":5.8},
        "upgrades": 5
    }
}

# 1) Build a Bennett factory focused on healing (combo="Q")
bennett_factory = build_character_factory(
    name="Bennett",
    weapon_name="Mistsplitter",
    artifacts=artifacts,
    aa_level=1,
    skill_level=1,
    burst_level=1,
    combo="Q"
)

# 2) Instantiate & sync Bennett
bennett = bennett_factory()
bennett.sync_stats(bennett.compute_total_stats())

# 3) Load multipliers (with your Bennett AA/Skill/Burst entries)
multipliers = load_talent_multipliers("../data/talent_multipliers.csv")

# 4) Define a healing metric: total heal over 12 s from one Q
def bennett_heal_metric(char):
    he = HealEffect(
        character   = char,
        multipliers = multipliers,
        skill_key   = "Burst",
        level       = char.burst_level,
        interval    = 2.0,    # Fantastic Voyage ticks every 2 s
        duration    = 12.0    # lasts 12 s
    )
    return he.total_heal()

# 5) Run the healing‐focused simulation *with* the new stat‐threshold policy
resin_budget = 3000
healer_results = simulate_with_metric(
    character        = bennett,
    resin_budget     = resin_budget,
    multipliers      = multipliers,
    metric_fn        = bennett_heal_metric,
    artifact_runner  = simulate_artifact_farm_heal,
    talent_runner    = simulate_talent_farm_heal,
    domain_level     = 4,
    policy           = "stat_threshold_then_swap",
    primary          = "talent",                   # farm talents first
    threshold_stats  = {"burst_level": 9}          # until Burst ≥ 9
)

# 6) Print summaries
final = healer_results[-1]
print(f"Total Healing after {resin_budget} resin: {final['metric_after']:.0f}")
print(f"Healing Gain per resin          : {final['metric_gain']/resin_budget:.3f}")

# 6b) Print Bennett's final artifacts & stats
print("\nFinal Artifacts on Bennett:")
for slot, art in bennett.artifacts.items():
    print(f"  {slot}: {art}")

stats = bennett.compute_total_stats()
print("\nFinal Bennett Stats:")
print(f"  Base HP            : {stats['base_hp']:.0f}")
print(f"  Weapon ATK         : {stats['weapon_base_atk']:.0f}")
print(f"  Base ATK           : {stats['base_atk']:.0f}")
print(f"  Base DEF           : {stats['base_def']:.0f}")
print(f"  Total HP           : {stats['total_hp']:.2f}")
print(f"  Total ATK          : {stats['total_atk']:.2f}")
print(f"  Total DEF          : {stats['total_def']:.2f}")
print(f"  Crit Rate          : {stats['crit_rate']:.2%}")
print(f"  Crit DMG           : {stats['crit_dmg']:.2%}")
print(f"  Energy Recharge    : {stats['energy_recharge']*100:.2f}%")
print(f"  Healing Bonus      : {stats['healing_bonus']*100:.2f}%")
print(f"Talent Levels     : AA Lv. {bennett.aa_level}, "
      f"Skill Lv. {bennett.skill_level}, Burst Lv. {bennett.burst_level}")

# If you want to see how much HP% bonus you got from artifacts:
hp_pct = stats['percent_bonus']['hp'] * 100
print(f"  Artifact HP% bonus : {hp_pct:.2f}%")

# Optional: see each step
df = pd.DataFrame(healer_results)
print(df[["activity","metric_before","metric_after","metric_gain","resin_spent"]])
```

```python
import pandas as pd

# 1) (Re)build & sync your Bennett  
bennett = build_character_factory(
    name="Bennett",
    weapon_name="Harbinger of Dawn",
    artifacts=None,   # start with empty slots
    aa_level=1,
    skill_level=1,
    burst_level=1,
    combo="Q"
)()
bennett.sync_stats(bennett.compute_total_stats())

# 2) multipliers & metric_fn stay the same
multipliers = load_talent_multipliers("../data/talent_multipliers.csv")
def bennett_heal_metric(char):
    he = HealEffect(char, multipliers, "Burst", char.burst_level, interval=2.0, duration=12.0)
    return he.total_heal()

total_resin = 3000

# ── PHASE 1: Talent‐only → Burst ≥ 9 ──
phase1 = simulate_with_metric(
    character       = bennett,
    resin_budget    = total_resin,
    multipliers     = multipliers,
    metric_fn       = bennett_heal_metric,
    artifact_runner = simulate_artifact_farm_heal,  # not used here
    talent_runner   = simulate_talent_farm_heal,
    policy          = "stat_threshold_then_swap",
    primary         = "talent",
    threshold_stats = {"burst_level": 9},
)

used1   = sum(r["resin_spent"] for r in phase1)
remaining = total_resin - used1

# Phase 2: farm artifacts to hit ER ≥ 2.20, *ignoring* healing
def er_metric(char):
    return char.compute_total_stats()["energy_recharge"]

phase2 = simulate_with_metric(
    character       = bennett,
    resin_budget    = remaining,
    multipliers     = multipliers,
    metric_fn       = er_metric,                  # ← switch to ER metric
    artifact_runner = simulate_artifact_farm_heal,  
    talent_runner   = simulate_talent_farm_heal,   # not used in artifact_only
    policy          = "artifact_only"
)

used2    = sum(r["resin_spent"] for r in phase2)
remaining = remaining - used2

# Phase 3: now go back to healing-greedy with what's left
phase3 = simulate_with_metric(
    character       = bennett,
    resin_budget    = remaining,
    multipliers     = multipliers,
    metric_fn       = bennett_heal_metric,
    artifact_runner = simulate_artifact_farm_heal,
    talent_runner   = simulate_talent_farm_heal,
    policy          = "greedy_swap",
)

# Collate & inspect:
all_steps = phase1 + phase2 + phase3
final     = all_steps[-1]

print(f"Total healing after {total_resin} resin: {final['metric_after']:.0f}")
print(f"Healing gain per resin point : {final['metric_gain']/total_resin:.3f}")

print("\nFinal Talent Levels:")
print(f"  AA:    Lv. {bennett.aa_level}")
print(f"  Skill: Lv. {bennett.skill_level}")
print(f"  Burst: Lv. {bennett.burst_level}")

print("\nFinal Bennett Stats:")
stats = bennett.compute_total_stats()
print(f"  Total HP        : {stats['total_hp']:.2f}")
print(f"  Energy Recharge : {stats['energy_recharge']*100:.1f}%")
print(f"  Healing Bonus   : {stats['healing_bonus']*100:.1f}%")

print("\nFinal Artifacts:")
for slot, art in bennett.artifacts.items():
    print(f"  {slot}: {art}")

print("\nStep-by-step:")
df = pd.DataFrame(all_steps)
print(df[["activity","metric_before","metric_after","metric_gain","resin_spent"]])
```

```python
import pandas as pd
from core.utils   import build_character_factory
from core.talents import load_talent_multipliers
from simulation.simulator import (
    simulate_with_metric,
    simulate_artifact_farm_heal,
    simulate_talent_farm_heal
)

# 1) Build Bennett (your usual setup)
artifacts={                         # or your real artifact dict
        "Flower":  {"main_stat":"Flat HP",        "main_stat_value":4780, "substats":{},"upgrades":5},
        "Feather": {"main_stat":"Flat ATK",       "main_stat_value":311,  "substats":{},"upgrades":5},
        "Sands":   {"main_stat":"Energy Recharge%", "main_stat_value":55.1, "substats":{},"upgrades":5},
        "Goblet":  {"main_stat":"HP%",            "main_stat_value":46.6, "substats":{},"upgrades":5},
        "Circlet": {"main_stat":"Healing Bonus%", "main_stat_value":20.0, "substats":{},"upgrades":5},
    }
bennett = build_character_factory(
    name="Bennett",
    weapon_name="Mistsplitter",
    artifacts=artifacts,
    aa_level=1,
    skill_level=1,
    burst_level=1,
    combo="Q"    # we only care about his Burst buff
)()
bennett.sync_stats(bennett.compute_total_stats())

# 2) Load your talent multipliers (must include Bennett’s Burst → TeamBuff%)
multipliers = load_talent_multipliers("../data/talent_multipliers.csv")

# 3) Define “how big is his ATK-buff” at a given state
def bennett_flat_atk_metric(char):
    lvl = char.burst_level
    entry = multipliers["Bennett"]["Burst"][lvl].get("InspirationBuff", {"value":0})
    buff_pct = entry["value"] / 100.0
    # compute Bennett’s base ATK + weapon ATK
    base_atk = char.base_stats["atk"] + (char.weapon.base_atk if char.weapon else 0)
    return base_atk * buff_pct


# 4) Run simulate_with_metric **only farming talents** for 3 000 resin
buff_results = simulate_with_metric(
    character       = bennett,
    resin_budget    = 3000,
    multipliers     = multipliers,
    metric_fn       = bennett_flat_atk_metric,   # <— flat ATK now
    artifact_runner = simulate_artifact_farm_heal,
    talent_runner   = simulate_talent_farm_heal,
    policy          = "talent_only"
)

final = buff_results[-1]
flat_bonus = final["metric_after"]
print(f"→ Final flat ATK bonus after 3000 resin: {flat_bonus:.1f} ATK")
print(f"→ Flat ATK gain per resin point        : {flat_bonus/3000:.4f} ATK")

tal = final["talent_levels"]
print(f"→ Final Talent Levels: AA Lv. {tal['AA']}, Skill Lv. {tal['Skill']}, Burst Lv. {tal['Burst']}")
```

```python
import matplotlib.pyplot as plt
import copy
from simulation.policies import MinMaxerPolicy, PlannerPolicy
from simulation.simulator import simulate_with_policy

# Step 1: Set up base character once
base_char = copy.deepcopy(characters[0])
base_char.crit_rate = 0.5
base_char.crit_dmg = 1.0
base_char.dmg_bonus = 0.2
base_char.talent_multiplier = 2.0
base_char.talent_level = 6
base_char.book_inventory = {
    'Teachings': 0,
    'Guides': 0,
    'Philosophies': 0
}

# Step 2: Create separate copies for each persona
char_mm = copy.deepcopy(base_char)
char_pl = copy.deepcopy(base_char)

# Step 3: Run simulations
mm_results = simulate_with_policy(char_mm, resin_budget=1200, policy=MinMaxerPolicy())
pl_results = simulate_with_policy(char_pl, resin_budget=1200, policy=PlannerPolicy())

# ✅ Step 3.5: Extend Min-Maxer curve to match planner's resin span
if mm_results:
    final_mm_dps = mm_results[-1]['dps']
    last_resin_spent = 1200 - mm_results[-1]['resin_remaining']
else:
    final_mm_dps = base_char.expected_damage_output(
        talent_multiplier=base_char.talent_multiplier,
        crit_rate=base_char.crit_rate,
        crit_dmg=base_char.crit_dmg,
        dmg_bonus=base_char.dmg_bonus,
        enemy_level=90,
        enemy_resistance=0.1
    )
    last_resin_spent = 0

if pl_results:
    max_resin_spent = 1200 - pl_results[-1]['resin_remaining']
    resin_step = 20
    for resin in range(last_resin_spent + resin_step, max_resin_spent + resin_step, resin_step):
        mm_results.append({
            'resin_remaining': 1200 - resin,
            'dps': final_mm_dps
        })

# Step 4: Plot results
print(f"Min-Maxer results: {len(mm_results)}")
print(f"Planner results: {len(pl_results)}")

plt.figure(figsize=(10, 5))
plt.plot(
    [1200 - r['resin_remaining'] for r in mm_results],
    [r['dps'] for r in mm_results],
    label='Min-Maxer'
)
plt.plot(
    [1200 - r['resin_remaining'] for r in pl_results],
    [r['dps'] for r in pl_results],
    label='Planner'
)
plt.xlabel("Resin Spent")
plt.ylabel("Expected Damage")
plt.title("Persona DPS Progression vs Resin Spent")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
